```python
# Agrupando por Payment Method por Purchase Amount (USD)
purchase_by_payment = df.groupby('Payment Method')['Purchase Amount (USD)'].sum()

# Plotando
plt.figure(figsize=(5, 5))
purchase_by_payment.plot(kind='bar', color='skyblue')
plt.xlabel('Payment Method')
plt.ylabel('Purchase Amount (USD)')
plt.title('Purchase Amount (USD) by Payment Method')
plt.xticks(rotation=45)
plt.show()
```

```python
# Convertendo a coluna 'Date Purchase' para o tipo datetime
df['Date Purchase'] = pd.to_datetime(df['Date Purchase'], errors='coerce')

# Criando a coluna 'Month_Year' com o formato correto
df['Month_Year'] = df['Date Purchase'].dt.strftime('%B %Y')

# Agrupando por mês e ano por Purchase Amount (USD) para cada mês
purchase_by_month_year = df.groupby('Month_Year')['Purchase Amount (USD)'].sum()

# Reindexando purchase_by_month_year para garantir ordem cronológica
purchase_by_month_year = purchase_by_month_year.reindex(pd.to_datetime(purchase_by_month_year.index, format='%B %Y').sort_values().strftime('%B %Y'))

# Plotando o gráfico de barras com ordem cronológica
plt.figure(figsize=(12, 6))
purchase_by_month_year.plot(kind='bar', color='skyblue')
plt.xlabel('Month and Year')
plt.ylabel('Purchase Amount (USD)')
plt.title('Purchase Amount (USD) by Month and Year')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
